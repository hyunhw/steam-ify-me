#### Absolute number of genres owned should not be used, the *fraction* should be used

```python
predictors = ['recommended',
    'num_owned_games', 
    'game_play_time',
    'lag_month',
    'dlc_price',
    'game_price',
    'Accounting', 'Action', 'Adventure', 'Animation&Modeling',
    'AudioProduction', 'Casual', 'Design&Illustration', 'EarlyAccess',
    'Education', 'FreetoPlay', 'Indie', 'MassivelyMultiplayer',
    'PhotoEditing', 'RPG', 'Racing', 'Simulation', 'SoftwareTraining',
    'Sports', 'Strategy', 'Utilities', 'VideoProduction', 'WebPublishing']
genres = ['Accounting', 'Action', 'Adventure', 'Animation&Modeling',
    'AudioProduction', 'Casual', 'Design&Illustration', 'EarlyAccess',
    'Education', 'FreetoPlay', 'Indie', 'MassivelyMultiplayer',
    'PhotoEditing', 'RPG', 'Racing', 'Simulation', 'SoftwareTraining',
    'Sports', 'Strategy', 'Utilities', 'VideoProduction', 'WebPublishing']
x = X[predictors]
x.head()
```

```python
for column in x.columns:
    if column not in ['recommended', 'num_owned_games', 'game_play_time','lag_month','dlc_price','game_price']:
        #print(column)
        x[column] = x[column] / (x['num_owned_games'] * 22)
```

#### people were more likely to recommend a DLC when:

- the user owned a lot of popular genres like "Action", "Adventure", "Early Access", "Free to Play", "MMO", "Racing", "Sim", "Sports"
